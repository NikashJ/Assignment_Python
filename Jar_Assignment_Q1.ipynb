```python
# Load the datasets:-
orders = pd.read_csv(r"C:\Users\91861\Desktop\Jar\List_of_Orders_55FFC79CF8.csv")
order_details = pd.read_csv(r"C:\Users\91861\Desktop\Jar\Order_Details_19795F61CF.csv")
sales_target = pd.read_csv(r"C:\Users\91861\Desktop\Jar\Sales_target_DD2E9B96A0.csv")
```

```python
# Merge the orders and order details datasets on 'Order ID'
merged_data = pd.merge(order_details, orders, on="Order ID")
```

```python
merged_data.head(10)
```

```python
# Calculate total sales per row (Quantity × Price)
merged_data["Total Sales"] = merged_data["Quantity"] * merged_data["Amount"]
```

```python
# Group by 'Category' and sum 'Total Sales'
category_sales = merged_data.groupby("Category")["Total Sales"].sum().reset_index()
```

```python
print(merged_data["Total Sales"])
```

```python
print(category_sales)
```

```python
# Group by 'Category' and calculate total profit, total sales, and total orders
category = merged_data.groupby("Category").agg({
    "Profit": "sum",  # Total Profit
    "Total Sales": "sum",  # Total Sales
    "Order ID": "count"  # Total Orders (Order count per category)
}).reset_index()
```

```python

# Rename columns
category.rename(columns={"Profit": "Total_Profit", "Total Sales": "Total_Sales", "Order ID": "Total_Orders"}, inplace=True)
```

```python
# Convert to numeric to avoid errors
category["Total_Profit"] = pd.to_numeric(category["Total_Profit"])
category["Total_Orders"] = pd.to_numeric(category["Total_Orders"])
category["Total_Sales"] = pd.to_numeric(category["Total_Sales"])
```

```python
# Calculate Average Profit per Order
category["Avg_Profit_per_Order"] = category["Total_Profit"] / category["Total_Orders"]
```

```python
# Calculate Profit Margin (Profit as a percentage of Sales)
category["Profit_Margin"] = (category["Total_Profit"] / category["Total_Sales"]) * 100
```

```python
print(category)
```

```python
# Find top-performing category (highest profit)
top_category = category.sort_values(by="Total_Profit", ascending=False).head(1)
```

```python
# Find underperforming category (lowest profit)
underperforming_category = category.sort_values(by="Total_Profit", ascending=True).head(1)
```

```python
print("Top-Performing Category:\n", top_category)
print("\nUnderperforming Category:\n", underperforming_category)
```

```python
# Find category with highest profit margin
top_margin_category = category.sort_values(by="Profit_Margin", ascending=False).head(1)
```

```python
# Find category with lowest profit margin
low_margin_category = category.sort_values(by="Profit_Margin", ascending=True).head(1)
```

```python
print("\nCategory with Highest Profit Margin:\n", top_margin_category)
print("\nCategory with Lowest Profit Margin:\n", low_margin_category)
```

```python
# Counting the top 5
top_states = orders["State"].value_counts().head(5).index

# Filter orders for these top states
filtered_ord = orders[orders["State"].isin(top_states)]

# Merge with order details to get sales and profit data
state_sales = pd.merge(filtered_ord, order_details, on="Order ID")

# Aggregate total sales and average profit by state
state_performance = state_sales.groupby("State").agg(
    Total_Sales=("Amount", "sum"),
    Average_Profit=("Profit", "mean"),
    Order_Count=("Order ID", "count")
).reset_index()

# Sort by Order Count to ensure highest first
state_performance = state_performance.sort_values(by="Order_Count", ascending=False)

# Display results
state_performance
```
